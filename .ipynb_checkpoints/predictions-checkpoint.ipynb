```python
import numpy as np

# Charger vos donnÃ©es de test dans X_test et Y_test
# Assurez-vous que X_test et Y_test ont la mÃªme longueur et sont correctement alignÃ©s

# Faire les prÃ©dictions sur les donnÃ©es de test
y_pred = model.predict(X_test)

# Obtenir les classes prÃ©dites en prenant l'indice de la valeur maximale dans chaque prÃ©diction
y_pred_classes = np.argmax(y_pred, axis=1)

# CrÃ©er un tableau pour stocker les rÃ©sultats combinÃ©s
results = np.zeros((len(Y_test), 3), dtype=int)  # 3 colonnes pour lose, draw, win

# Remplir le tableau avec les rÃ©sultats rÃ©els et prÃ©dits
results[:, 0] = Y_test  # RÃ©sultats rÃ©els dans la premiÃ¨re colonne
results[:, 1] = y_pred_classes  # RÃ©sultats prÃ©dits dans la deuxiÃ¨me colonne

# Calculer les rÃ©sultats de prÃ©diction en utilisant les classes prÃ©dites
# 0 pour lose, 1 pour draw, 2 pour win
results[:, 2] = np.where(y_pred_classes == 0, 0,  # lose
                        np.where(y_pred_classes == 1, 1,  # draw
                                 2))  # win

# Afficher les rÃ©sultats sous forme de tableau
print("RÃ©sultats rÃ©els vs prÃ©dits:")
print("RÃ©alitÃ© | PrÃ©diction | PrÃ©diction (0: lose, 1: draw, 2: win)")
for i in range(len(Y_test)):
    print("{:^8} | {:^11} | {:^26}".format(results[i, 0], results[i, 1], results[i, 2]))
```
